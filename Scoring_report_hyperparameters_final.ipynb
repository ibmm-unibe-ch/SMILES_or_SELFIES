# Choose hyperparameters
## Load data

```python
import pandas as pd
import os
from constants import TASK_MODEL_PATH, TOKENIZER_SUFFIXES, MOLNET_DIRECTORY, PROJECT_PATH,  DESCRIPTORS
from glob import glob
from pathlib import Path
pd.set_option('display.max_rows', 500)

HYPERPARAMETER_PATH = Path("/data/jgut/SoS_models/")

classification_scores = []
regression_scores = []
suffixes = []
for suffix in TOKENIZER_SUFFIXES:
    suffixes.extend([f"{suffix}_bart",f"{suffix}_roberta",suffix])
for tokenizer_suffix in suffixes:
    for task in MOLNET_DIRECTORY.keys():
        task_tokenizer_path = HYPERPARAMETER_PATH/task/tokenizer_suffix
        print(task_tokenizer_path)
        if task_tokenizer_path.exists():
            for hyperparameter_path in glob(str(task_tokenizer_path) + "/*based_norm", recursive=True):
                scores_path = hyperparameter_path+"/scores.csv"
                print(scores_path)
                if not Path(scores_path).is_file():
                    continue
                new_score_df = pd.read_csv(scores_path)
                if list(new_score_df.task_type)[0] == "classification":
                    classification_scores.append(new_score_df)
                else:
                    regression_scores.append(new_score_df)

regression_scores = pd.concat(regression_scores, axis = 0, sort = False)
classification_scores = pd.concat(classification_scores, axis = 0, sort = False)

for table in [regression_scores, classification_scores]:
    table.reset_index(inplace=True)
    for row in table[table.embedding.isna()][["embedding","tokenizer","dataset", "architecture"]].itertuples():
        embedding = row.tokenizer.split("_")[0]
        tokenizer = row.tokenizer.split("_")[1]
        dataset = row.tokenizer.split("_")[2]
        table.at[row.Index, 'architecture'] = "bart"
        table.at[row.Index, 'embedding'] = embedding
        table.at[row.Index, 'tokenizer'] = tokenizer
        table.at[row.Index, 'dataset'] = dataset
classification_scores = classification_scores.drop(["index", "Unnamed: 0", "model_size","data_type","seed", "task_type"], axis="columns")
regression_scores = regression_scores.drop(["index", "Unnamed: 0", "model_size","data_type","task_type"], axis="columns")
```
